```python
import pandas as pd

offset = 50
savedtracks = sp.current_user_saved_tracks(limit = 50)['items']
while offset < 2000:
    
    savedtracks2 =sp.current_user_saved_tracks(limit = 50,offset=offset)['items']
    for track in savedtracks2:
        savedtracks.append(track)
    offset += 50
    
```

```python
a = {}

for track in savedtracks:
    id = track['track']['id']
    
    features = sp.audio_features(id)[0]

    a[track['track']['name']] =    {"popularity" : track['track']['popularity'] ,
                                 "energy" :features['energy'],
                                 "id" : id,
                                 "liveness" :features['liveness'],
                                 "danceability" :features['danceability']     ,
                                 "instrumentalness" : features['instrumentalness'] }
```

```python
user =  'tonyryanworldwide'

df = pd.DataFrame.from_dict(a,orient = "index").sort_values(['energy','popularity'], ascending = False)
highenergy = df[df['energy']  > .90]
danceability = df[(df['danceability']  > .80) & (df['energy'] > .8)]

# trackstoadd = highenergy.sort_values('popularity')[201:]['id'].tolist()
trackstoadd = danceability[0:100]['id'].tolist()

playlist_id = sp.user_playlist_create(user=username, name ='JustDance',public = False,description = "Dance")['id']
# playlist_id = sp.user_playlist_create(user='tonyryanworldwide', name ='High Energy',public = False,description = "Fuck shit up")['id']

sp.user_playlist_add_tracks(user,playlist_id,tracks = trackstoadd)
```
